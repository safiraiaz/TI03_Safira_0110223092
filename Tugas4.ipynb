```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/DataMining/Groceries_dataset.csv', names = ['itemDescription'], sep = ',')
df.head()
```

```python
df.values
```

```python
data = list(df["itemDescription"].apply(lambda x:x.split(",") ))
data
```

```python
from mlxtend.preprocessing import TransactionEncoder
te = TransactionEncoder()

data_encoded = te.fit(data).transform(data)

df = pd.DataFrame(data_encoded, columns=te.columns_)
df = df.astype(int)
df
```

```python
# Convert 'itemDescription' to a boolean format with one-hot encoding for each item
df_boolean = df.copy()
df_boolean = pd.get_dummies(df_boolean, columns=['itemDescription'], prefix='', prefix_sep='')

# Display the first few rows of the modified DataFrame
df_boolean.head()
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
frequent_itemsets = apriori(df_boolean, min_support=0.01, use_colnames=True)
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.1)
rules
```
