```python
# import dan tampilkan dataset
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('/content/drive/MyDrive/DataMining/tested.csv')
print(data)
```

```python
# melihat datatype dan spesifik pada dataset
data.info()
```

```python
# Membuat DataFrame
df_titanic = pd.DataFrame(data)
df_titanic
```

```python
# Menghitung rata-rata Survived
mean_Survived = df_titanic['Survived'].mean()
mean_Survived
```

```python
# Menghitung nilai minimum dan maksimum pada kolom Survived
min_Survived = data['Survived'].min()
max_Survived = data['Survived'].max()

print(f'Nilai maksimum Survived: {max_Survived}')
```

```python
# mengecek apakah ada missing values
data.isnull().sum()
```

```python
# Hapus data yang tidak diperlukan karena kesuluruhan data memiliki nilai kosong
columns_to_drop = ['Age', 'Fare', 'Cabin']
data.drop(columns=columns_to_drop, inplace=True)
data
```

```python
# menjadikan PassengerId sebagai index agar memudahkan mengambil sebuah data
data.set_index('PassengerId', inplace=True)
data
```

```python
# Renaming columns
data.rename(columns={
    'Survived': 'Selamat',
    'Pclass': 'Kelas Tiket',
    'Name': 'Nama',
    'Sex': 'Jenis Kelamin',
    'SibSp': 'Saudara/Pasangan',
    'Parch': 'Orang Tua/Anak',
    'Ticket': 'Tiket',
    'Embarked': 'Embarked',
}, inplace=True)

#Tampilan DataFrame setelah perubahan nama kolom
data
```

```python
# Visualisasi 1
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='Selamat', hue='Jenis Kelamin', kde=True)
plt.title('Distribusi Penumpang Selamat Berdasarkan Jenis Kelamin')
plt.xlabel('Selamat')
plt.ylabel('Jumlah Penumpang')
plt.show()
```

```python
# Visualisasi 2
plt.figure(figsize=(10, 6))
sns.barplot(data=data, x='Kelas Tiket', y='Selamat', hue='Jenis Kelamin')
plt.title('Tingkat Kelangsungan Hidup Berdasarkan Kelas Penumpang dan Jenis Kelamin')
plt.xlabel('Kelas Tiket')
plt.ylabel('Tingkat Kelangsungan Hidup')
plt.show()
```

```python
# Visualisasi 3: Violin plot untuk melihat distribusi nama berdasarkan kelas penumpang (Pclass)
plt.figure(figsize=(8, 6))
sns.violinplot(x='Kelas Tiket', y='Nama', data=data, palette='muted')
plt.title('Nama Distribusi dengan Kelas Tiket', fontsize=15)
plt.xlabel('Kelas Tiket', fontsize=12)
plt.ylabel('Nama', fontsize=12)
plt.grid(True)
plt.show()
```

```python
# Melakukan normalisasi pada data numerik
from sklearn.preprocessing import MinMaxScaler

# Memilih kolom-kolom numerik yang akan dinormalisasi
numeric_cols = ['Kelas Tiket', 'Selamat', 'Saudara/Pasangan', 'Orang Tua/Anak']

# Inisialisasi MinMaxScaler
scaler = MinMaxScaler()

# Melakukan fit dan transformasi pada data numerik
df_normalized = data.copy()
df_normalized[numeric_cols] = scaler.fit_transform(data[numeric_cols])

# Menampilkan data setelah dinormalisasi
df_normalized.head()
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Pilih Kolom Numerik untuk Transformasi
numerical_cols = ['Kelas Tiket', 'Orang Tua/Anak']

# Buat objek StandardScaler
scaler = StandardScaler()

# Terapkan Z-Score Transformation
data[numerical_cols] = scaler.fit_transform(data[numerical_cols])

# Lihat Hasilnya
print(data[numerical_cols].head())
```

```python
import pandas as pd

# Pilih Kolom Numerik untuk Transformasi
numerical_cols = ['Saudara/Pasangan', 'Orang Tua/Anak']

# Hitung j (jumlah digit terpanjang)
j = data[numerical_cols].apply(lambda x: len(str(int(x.max())))).max()

# Terapkan Decimal Scaling
data[numerical_cols] = data[numerical_cols] / (10 ** j)

# Lihat Hasilnya
print(data[numerical_cols].head())
```
